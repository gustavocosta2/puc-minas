```python
# Função que classifica o tipo com base nas condições de pressão alta, doenças CVD e outras condições.
def determinar_tipo(row):
    # Verifica se todas as colunas relevantes estão iguais a 2
    todas_igual_2 = all(row[col] == 2 for col in [
        'Q03001',  # DIABETES
        'Q060',    # COLESTEROL ALTO
        'Q068',    # AVC
        'Q074',    # ASMA
        'Q079',    # ARTRITE
        'Q088',    # DORT
        'Q092',    # DEPRESSÃO
        'Q11006',  # DOENÇAS MENTAIS
        'Q11604',  # DOENÇAS CRÔNICAS
        'Q120',    # CÂNCER
        'Q124',    # INSUFICIÊNCIA RENAL
        'K045',    # CATARATA
        'Q06306',   # DOENÇAS CARDIOVASCULARES
        'Q00201'
    ])
    
    if row['Q00201'] == 1 and row['Q06306'] ==  1 and row["Q003"] == row["C008"]:
        return 1  # Tipo 1: Pressão Alta e Doenças CVD = 1.
    elif row['Q00201'] == 1 and row['Q06306'] == 2 and row["Q003"] == row["C008"]:
        return 2  # Tipo 2: Apenas Pressão Alta = 1.
    elif row['Q00201'] == 2 and row['Q06306'] == 2 and todas_igual_2:
        return 3  # Tipo 3: Pessoas sem diagnóstico.
    else:
        return None  # Nenhuma das condições especificadas

# Aplica a função para cada linha do DataFrame e cria a nova coluna 'Tipo'
df['Classe'] = df.apply(determinar_tipo, axis=1)
```

## **1.2. Conjunto 1: Pessoas Saudáveis**

Por definição, as pessoas saudáveis serão todas aquelas que responderam diferente de Sim para os diagnósticos e que não tenham como resposta nulo.

```python
def salvar_conjunto(dataframe):
    diretorio = "../dados"

    if not os.path.exists(diretorio):
        os.makedirs(diretorio)

    #Pergunta ao usuário do código o nome do arquivo a ser salvo.
    nome = input("Digite o nome do arquivo a ser salvo: ")
    nome += ".csv"

    caminho_arquivo = os.path.join(diretorio, nome)
    
    #Salvar o DataFrame em .csv
    dataframe.to_csv(caminho_arquivo, index=False)

    print(f"Arquivo salvo em: {caminho_arquivo}")

    
```

```python
# Encontrando o total de pessoas SAUDÁVEIS, isto é, sem NENHUMA comorbidade.
df_saudaveis = df.loc[
    (df['Q03001'] != 1) & #diabetes
    (df['Q060'] != 1)   & #colesterol alto
    (df['Q068'] != 1)   & #avc
    (df['Q074'] != 1)   & #asma
    (df['Q079'] != 1)   & #artrite
    (df['Q088'] != 1)   & #dort
    (df['Q092'] != 1)   & #depressao
    (df['Q11006'] != 1) & #doencas mentais
    (df['Q11604'] != 1) & #doenca cronica pulmonar
    (df['Q120'] != 1)   & #cancer
    (df['Q124'] != 1)   & #insuficiencia renal
    (df['K047'] != 1)   & #catarata
    (df['Q00201'] != 1) & #pressao alta
    (df['Q06306'] != 1)   # doenças cardiovasculares
]
```

```python
df_saudaveis.shape
```

```python
salvar_conjunto(df_saudaveis)
```
